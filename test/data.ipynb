```python
import httpx
import json

base_url = "https://api.imf.org/external/sdmx/3.0"

def get_dataset(endpoints):
    # Enable follow_redirects=True to handle the IMF 302 routing shift smoothly
    with httpx.Client(timeout=httpx.Timeout(45.0, connect=15.0), follow_redirects=True) as client:
        r = client.get(
            f'{base_url}{endpoints}', 
            headers={"Accept": "application/json"}
        )
        r.raise_for_status()
        data = r.json()
        
        # 1. Build the dictionary mapping flow IDs to Names
        flows = data["data"]["dataflows"]
        dataflows = {f["id"]: f["name"] for f in flows}
        
        # 2. Save the full API raw response
        with open('/mnt/D06A89C26A89A5B6/projects/portfolio/Hermes/test/data1.json', 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=4)
            
        # 3. Save your custom filtered dictionary directly to data.json
        with open('/mnt/D06A89C26A89A5B6/projects/portfolio/Hermes/test/data.json', 'w', encoding='utf-8') as f:
            json.dump(dataflows, f, indent=4, ensure_ascii=False)

get_dataset('/structure/dataflow')
```
